```python
#Handling missing data in python using pandas library
## now using pandas library, we are going to use dropna
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
df=pd.read_csv('C:\\Users\\Abdul Nasir\\Desktop\\Jupyter_Notebook\\fao_lib.csv')
# this will only return the row which is filled with values or which are not filled with NaN
df.dropna()
```

```python
# now return those coloumn having no NaN field
df.dropna(axis=1)
```

```python
# Handling Missing data in dataframe
# now drop row having NaN Values in, for this purpose we are going to used how='any' parameter of dropna() method
df.dropna(how='any')
```

```python
# now trying another ways using two parameter thatis 'how' and 'axis' for handling missing data in dataframe
df.dropna(axis=1, how='any')
```

```python
# now we can handle missing data in another ways as
## how='all' parameter of dropna() method, let us check the data behavior
### this will return rows and coloumn of data frame having atleast one not null values in row i.e NaN 
df.dropna(how='all')
```

```python
# Missing data frame handling using thresh(Threshold data) for pandas method dropna()
# thresh values may be 1,2,3,4,5 mean any scaler values
df.dropna(thresh=1)
```

```python
# Missing data frame handling using thresh(Threshold data) for pandas method dropna()
# thresh values may be 1,2,3,4,5 mean any scaler values
df.dropna(thresh=3)
```

```python
# now another parameter of dropna() method i.e subset=[] for missing data handling
df.dropna(subset=['Note'])
```

```python
# now another parameter of dropna() method i.e subset=[] for missing data handling
# this will only return those rows and colomn havinf not NaN values
df.dropna(subset=['Note'])
```

```python
# now we are going to explore some other methods used for missing data handling in pandas
## method use to fill with your own values
# general Sytax for method fillna()
# dataframe.fillna()
# now let use it we are using method='ffill' which mean forward fill
df.fillna(method='ffill')
```

```python
# now we are going to explore some other methods used for missing data handling in pandas
## method use to fill with your own values
# general Sytax for method fillna()
# dataframe.fillna()
# now let use it we are using method='bfill' which mean bakcward fill
df.fillna(method='bfill')
```

```python
# now we are going to explore some other methods used for missing data handling in pandas
## method use to fill with your own values
# general Sytax for method fillna()
# dataframe.fillna()
# now let use it we are using method='pad' which mean bakcward fill
df.fillna(method='pad')
```

```python
# now we are going to explore some other methods used for missing data handling in pandas
## method use to fill with your own values
# general Sytax for method fillna()
# dataframe.fillna()
# now let use it we are using method='ffill' which mean bakcward fill and also use axis parameter. i.e axis=1
df.fillna(method='ffill', axis=0)
```

```python
# now we are going to explore some other methods used for missing data handling in pandas
## method use to fill with your own values
# general Sytax for method fillna()
# dataframe.fillna()
# now let use it we are using method='bfill' which mean bakcward fill and also use axis parameter. i.e axis=1
df.fillna(method='bfill', axis=1)
```

## import libraries
- seaborn automatically install these library
- numpy
- scipy
- pandas
- matplotlib

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
# load dataset
phool=sns.load_dataset("iris")
phool
#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.show()
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
# load dataset
phool=sns.load_dataset("iris")
#change size of figure
plt.figure(figsize=(8,6))
#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Nasir performing practice on line plot")
plt.show()
```
